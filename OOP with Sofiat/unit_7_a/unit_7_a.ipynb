## Attributes (aka Variables)

### 1. Instance variables

````python
class Student:
    
    # init method to initialise instance variables
    def __init__(self, first_name, last_name, lab):
        self.first_name = first_name        
        self.last_name = last_name
        self.lab = lab
````

<div class="alert alert-info"> 
    <b>NOTE:</b> To add line numbers to your cell, go to View then Toggle Line Numbers.    
</div>

**Line 1.** The class `Student` is created.

**Line 4.** The variables that all `Student` objects must have are defined in a method called `__init__()`. 
* `__init__()` is referred to as the **constructor** of the class. 
* Every time a new `Student` object is created, `__init__()` sets the initial state of the object by assigning the values of the objects variables. 
* In essence, Python uses the `__init__()` method to initialize each new instance of the class.

* You can give `__init__()` any number of parameters, but the first parameter will always be a variable called `self`. 
* When a new class instance is created, the instance is automatically passed to the `self` parameter in `__init__()` so that new variables can be defined on the object.
* In our `Student` class, `__init__()` method has three variables: `first_name`, `last_name`, and `lab`.

**Lines 5 - 7.** Here, we have three statements under the body of `__init__()` that uses the `self` variable.
* These variables: `self.first_name`, `self.last_name` and `self.lab` are referred to as the **instance variables**. 
* You MUST initialise all parameters within the `__init__()` method this way. 
* To access and update the value of an instance variable anywhere within the class, use `<self>.<name_of_instance_variable>`.
* Just as you can call a function multiple times with different parameters, the value of an instance variable is unique to a particular instance of the class. All `Student` objects have a `first_name`, `last_name` and `lab`, but the values for these variables will vary depending on the `Student` instance.
* Instance variables are to classes as global variables are to ordinary functions.

````python
# we create an instance from the Student class
 
student1 = Student("Jacob", "Liu", "LB10")
student1.first_name
````

### 2. Class variables

````python
class Student:
    
    # class variables    
    level = "undergraduate"
    total_students = 0
    
    # init method to initialise instance variables
    def __init__(self, first_name, last_name, lab):
        self.first_name = first_name        
        self.last_name = last_name
        self.lab = lab
        Student.total_students += 1
````

**Lines 4 - 5.** Variables `level` and `total_students` are defined outside of the `__init__()` method because we want them to have the same value for all instances created from the class. They are referred to as **class variables**.
* They must always be assigned an initial value
* When an instance of `Student` is created, variables `level` and `total_students` are automatically created and assigned to their initial (or updated) values.

**Line 12.** To access and update the value of a class variable anywhere within the class, use `<class_name>.<name_of_class_variable>`.

### Class variables Vs Instance variables

* Properties:
    * class variables: same value for every instance of the class;
    * instance variables: value is unique to a class instance
<br>

* Access and update within the class:
    * class variables: `<class_name>.<name_of_class_variable>`;
    * instance variables: `<self>.<name_of_instance_variable>`.

* Access and update outside the class:
    * for both: use `<object_name>.<name_of_variable>`.

## Methods

### 1. Instance methods

* They are functions that are defined inside the body of a class.
* They operate on the variables of a specific instance of a class. For example:

```python
a = "string"
a.upper()
```

* When we call the `.upper()` method on variable `a`, we are converting the content of `a` to upper case.

* They can only be called from an instance of that class. For example, we cannot access the `.upper()` method without first creating a string.

````python
a = "string"
a.upper()
````

````python
class Student:
    
    # class variables    
    level = "undergraduate"
    total_students = 0
    
    # init method to initialise instance variables
    def __init__(self, first_name, last_name, lab):
        self.first_name = first_name        
        self.last_name = last_name
        self.lab = lab
        Student.total_students += 1 # increment total number of students        
    
    # instance method to get full name
    def full_name(self):
        return f"{self.first_name} {self.last_name}"
    
    # instance method
    def mood(self):
        return f"{self.full_name()} enjoys Python programming!"
    
    # instance method to get email
    def get_email(self):
        full_name = f"{self.first_name.lower()}.{self.last_name.lower()}"
        return f"{full_name}@student.gla.ac.uk"
    
    # instance method to allow a student change lab
    def change_lab(self, new_lab):        
        self.lab = new_lab
````

* From line 15, we define instance methods using the `def` keyword (just as we would for functions): 
    * `full_name()`
    * `mood()`
    * `get_email()`
    * `change_lab()` 
* Similar to `__init__()`, the first parameter of an instance method is always `self`.
* You can also pass additional parameters to an instance method, but the scope would only be visible within this method.
* To use an instance method within the class: `<self>.<method_name()>`
* To use an instance method outside the class: `<object_name>.<method_name()>`

## 2. Class methods

* These are functions that are bound to the class and not to a specific object
* They are shared by all the objects created from the class
* The operate on class-level data without modifying individual objects

````python
class Student:
    
    # class variables    
    level = "undergraduate"
    total_students = 0
    all_students = [] # new class variable
    
    # init method to initialise instance variables
    def __init__(self, first_name, last_name, lab):
        self.first_name = first_name        
        self.last_name = last_name
        self.lab = lab
        Student.total_students += 1 
        # keep track of each Student object created from our class
        Student.all_students.append(self) 
    
    # instance method to get full name
    def full_name(self):
        return f"{self.first_name} {self.last_name}"
    
    # instance method
    def mood(self):
        return f"{self.full_name()} enjoys Python programming!"
    
    # instance method to get email
    def get_email(self):
        full_name = f"{self.first_name.lower()}.{self.last_name.lower()}"
        return f"{full_name}@student.gla.ac.uk"
    
    # instance method to allow a student change lab
    def change_lab(self, new_lab):        
        self.lab = new_lab
    
    # class method to get all students
    @classmethod
    def get_all_students(cls):
        # this returns an object representation of each Student created from our class
        return cls.all_students
    
     # class method to get students in the same lab
    @classmethod
    def get_students_in_same_lab(cls, lab):
        # for each student in the list
        # if the student's lab is same as the given lab
        # store their full name, which we extract from their ObjectName via the full_name method
        #same_lab = [stud.full_name() for stud in cls.all_students if stud.lab == lab]
        same_lab = []
        for stud in cls.all_students: # for each Student object stored in all_students
            if stud.lab == lab:
                same_lab.append(stud.full_name())
        return same_lab
````

* From line 35, we define class methods using `def` but with the `@classmethod` decorator to distinguish them from instance methods
    * `get_all_students`
    * `get_students_in_same_lab`
* The first parameter of a class method is `cls` (not `self`), since the class method will not be bound to a specific object.
* To use a class method within the class: `<cls>.<method_name()>`
* To use a class method outside the class: `<object_name>.<method_name()>`

## Displaying object representations with `__str__` and `__repr__`

In Python, `__str__` and `__repr__` are special methods that you can define in your class to control how instances of your class are represented as strings. They are used in different contexts:

* `__str__`: This method is called by the `str()` and the `print()` function. It returns a _human-readable_ string containing useful information about an instance of a class.

* `__repr__`: This method is called by the `repr()` built-in function. It is used whenever you evaluate an expression that returns an object (i.e., when you don't use `print`).


````python
class Student:
    
    # class variables    
    level = "undergraduate"
    total_students = 0
    all_students = [] # new class variable
    
    # init method to initialise instance variables
    def __init__(self, first_name, last_name, lab):
        self.first_name = first_name        
        self.last_name = last_name
        self.lab = lab
        Student.total_students += 1 
        # keep track of each Student object created from our class
        Student.all_students.append(self) 
    
    # instance method to get full name
    def full_name(self):
        return f"{self.first_name} {self.last_name}"
    
    # instance method
    def mood(self):
        return f"{self.full_name()} enjoys Python programming!"
    
    # instance method to get email
    def get_email(self):
        full_name = f"{self.first_name.lower()}.{self.last_name.lower()}"
        return f"{full_name}@student.gla.ac.uk"
    
    # instance method to allow a student change lab
    def change_lab(self, new_lab):        
        self.lab = new_lab
    
    # class method to get all students
    @classmethod
    def get_all_students(cls):
        # this returns an object representation of each Student created from our class
        return cls.all_students
    
     # class method to get students in the same lab
    @classmethod
    def get_students_in_same_lab(cls, lab):
        # for each student in the list
        # if the student's lab is same as the given lab
        # store their full name, which we extract from their ObjectName via the full_name method
        #same_lab = [stud.full_name() for stud in cls.all_students if stud.lab == lab]
        same_lab = []
        for stud in cls.all_students: # for each Student object stored in all_students
            if stud.lab == lab:
                same_lab.append(stud.full_name())
        return same_lab
    
    # string method to print object of Student class
#     def __str__(self):
#         return f"{self.first_name} {self.last_name} is in {self.lab}."

    # representation method
    def __repr__(self):
        return f"{self.first_name} ------> {self.last_name}"
````

````python
dict_ = {}
with open("file.txt") as f:
    # iterate through each line in f 
    # and start the index counter from 1
    for idx, line in enumerate(f, 1):
        first, last, lab =  line.strip().split()
        object_name = f"student{idx}" # student1, student2, etc...
        # using dict_, point each object name to an instance of the class Student
        dict_[object_name] = Student(first, last, lab)
````

**Key take-away:**

* If you implement `__str__` method, and you print an object, the `__str__` method will be called; otherwise, `__repr__` will be called if present.
* If you implement `__str__` only and you don't call the object with `print`, you will see the memory location (which is not readable to you or the user)
* `__repr__` provides a backup for `__str__`
* Best practice is to implement both `__repr__` and `__str__`, but if you can only write one, then implement `__repr__`.
