```python
#
# Import required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 
# Generate a Line Plot
x = np.linspace(0, 0.9, 12)
y = x**5 + x**4 + x**3 + 2.1*(x**2)
plt.plot(x, y)

plt.title("Changes in some random numbers")

plt.xlabel("Some random numbers")
plt.ylabel("Function of some random numbers")
```

```python
# 
# Scatter plot
# 
# student IDs
x = [5, 23, 23, 43, 2, 1, 34, 35, 6, 65]
# 
# Student scores
y = [500, 263, 293, 433, 221, 390, 434, 378, 290, 351]
plt.scatter(x, y)
```

```python
# 
# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
plt.bar(categories, values)
```

```python
# 
# Histogram
import numpy as np
data = np.random.normal(0, 1, 100)
plt.hist(data, bins=10)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
