```python
#
# Import required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 
# Generate a Line Plot
x = np.linspace(0, 0.9, 12)
y = x**5 + x**4 + x**3 + 2.1*(x**2)

plt.title("Changes in some random numbers")

plt.xlabel("Some random numbers")
plt.ylabel("Function of some random numbers")
line_plot = plt.plot(x, y)
```

```python
# 
# Scatter plot
# 
# student IDs
x = [5, 23, 23, 43, 2, 1, 34, 35, 6, 65]
# 
# Student scores
y = [500, 263, 293, 433, 221, 390, 434, 378, 290, 351]
scatter = plt.scatter(x, y)
```

```python
# 
# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
bar = plt.bar(categories, values)
```

```python
# 
# Histogram
import numpy as np
data = np.random.normal(0, 1, 100)
hist = plt.hist(data, bins=10)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 
# ASSIGNMENT: Create 4 sub plots
# 
# Create a figure object
fig = plt.figure(figsize=(10, 9))
# 
# Create 4 subplots (2 rows x 2 columns)
# 
# A line plot on Row 1, Col 1 
ax1 = fig.add_subplot(2, 2, 1)
# 
# Random data
x = np.linspace(0, 0.9, 12)
y = x**5 + x**4 + x**3 + 2.1*(x**2)
ax1.plot(x, y)
ax1.set_title("Changes in some random numbers on a Line Plot")
ax1.set_xlabel("Some random numbers")
ax1.set_ylabel("Function of some random numbers")
# 
# A Scatter Plot on Row 1, Col 2
ax2 = fig.add_subplot(2, 2, 2)
# 
# student IDs
std_ids = [5, 23, 23, 43, 2, 1, 34, 35, 6, 65]
# 
# Student scores
std_scores = [500, 263, 293, 433, 221, 390, 434, 378, 290, 351]
ax2.scatter(std_ids, std_scores)
ax2.set_title("Changes in some random numbers on a Scatter Plot")
ax2.set_xlabel("Some random numbers")
ax2.set_ylabel("Function of some random numbers")
# 
# A Bar Plot on Row 2, Col 1
ax3 = fig.add_subplot(2, 2, 3)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
ax3.bar(categories, values)
ax3.set_title("Values per Category on a Bar Plot")
ax3.set_xlabel("Categories")
ax3.set_ylabel("Values")
# 
# A Histogram on Row 2, Col 2
ax4 = fig.add_subplot(2, 2, 4)
data = np.random.normal(0, 1, 100)
ax4.hist(data, bins=10)
ax4.set_title("A Histogram showing the relationship between an arbitrary value and frequency")
ax4.set_xlabel("Value")
ax4.set_ylabel("Frequency")
```
