```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 


# 
# Get the fertility, population, gdp and life expectancy at birth data from world bank data for Kenya
# and merge it with the gapminder data for Kenya and visualize it
# 
# Set the parameters for the request
country_code = 'KE'
date_range = "1997:2022"
params = {
    'date': date_range,
    'format': 'json'
}
# 
# Get the Population Data
population_indicator_id = "SP.POP.TOTL"
# 
# Data source
url = f"https://api.worldbank.org/v2/country/{country_code}/indicator/{population_indicator_id}"
# 
# Send GET request to the remote resource
resp = requests.get(url, params)
# 
# Process returned data
if resp.status_code == 200:
    pop_data = resp.json()
    # 
    # Filter the returned data
    pop_data = pop_data[1]
    # 
    # Convert the list of dictionaries to DataFrame
    pop_df = pd.DataFrame(pop_data)
    # 
    # Clean up the DataFrame
    pop_df = pop_df[['date', 'value']]
    # 
    # Rename the columns
    pop_df = pop_df.rename(
        columns = {
            'date': 'year',
            'value': 'Population'
        }
    )
else:
    pass
pop_df
```

```python
# 
# Get Gapminder data
gapminder_df = pd.read_csv('../data/gapminder.csv')
gapminder_df
```
