```python
import requests
import pandas as pd
```

```python
# 
# Get the total value of goods and services produced by a UK in a year, measured in current US dollars from the world bank API
country_code = "GB"    # ISO code for the country
indicator_id = "NY.GDP.MKTP.CD" 
date_range = "2008:2022"
# 
# Data source
url = f"https://api.worldbank.org/v2/country/{country_code}/indicator/{indicator_id}"
# 
# Set the parameters for the request
params = {
    'date': date_range,
    'format': 'json'
}
# 
# Send GET request to the remote resource
resp = requests.get(url, params)
# 
# Process returned data
if resp.status_code == 200:
    dev_data = resp.json()
    # 
    # Filter the returned data
    gb_data = dev_data[1]
    # 
    # Convert the list of dictionaries to DataFrame
    gb_df = pd.DataFrame(gb_data)
    # 
    # Clean up the DataFrame
    gb_df = gb_df[['date', 'value']]
    # 
    # Rename the columns
    gb_df.rename(
        columns = {
            'date': 'Year',
            'value': 'GDP_USD'
        }
    )
else:
    pass
gb_df
```
