```python
import requests
import pandas as pd
```

```python
# 
# Write a function that takes the parameters for world bank data
def get_worldbank_data(country_code, indicator_id, indicator_name, date_range):
    """
        get_worldbank_data gets development indicator data for a given country

        Params:
            country_name
            indicator_id
            indicator_name
            date_range

        Returns:
            DataFrame of the indicator per year over a given year range
    """
    url = f"https://api.worldbank.org/v2/country/{country_code}/indicator/{indicator_id}"
    paramitas = {
        'date': date_range,
        'format': 'json'
    }
    # 
    # Send the GET request to the API.
    respons = requests.get(url, params = paramitas)
    # 
    # Process returned data.
    if respons.status_code == 200:
        indicator_data = respons.json()
        # 
        # Get the data from the second element of returned data
        indicator_data = indicator_data[1]
        # 
        # Convert the list of dictionaries to DataFrame
        indicator_data_df = pd.DataFrame(indicator_data)
        # 
        # Filter the columns to show for the DataFrame
        indicator_data_df = indicator_data_df[['date', 'value']]
        # 
        # Rename the columns
        indicator_data_df = indicator_data_df.rename(
            columns = {
                'date': 'year',
                'value': f"{indicator_name}"
            }
        )
        return indicator_data_df
    else:
        print("Eror getting indicator data", respons.status_code)
```

```python
# 
# Get Kenyan Population data from world bank data
ke_pop = get_worldbank_data("KE", "SP.POP.TOTL", "pop", "2008:2022")
ke_pop
```
