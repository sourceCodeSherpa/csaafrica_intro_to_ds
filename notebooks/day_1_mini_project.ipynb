```python
# 
# Import Pandas, Matplotlib, and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 
# Load a simple dataset called Iris from Seaborn
iris_dset = sns.load_dataset('iris')
# 
# Make a DataFrame out of the dataset above
iris_df = pd.DataFrame(iris_dset)
# 
# Show the Iris dataset as a DataFrame
iris_df
```

```python
# 
# Give a summary of the Iris Dataframe
iris_df.info()
```

```python
# 
# Show the total number for each unique value in the 'species' column
iris_df['species'].value_counts()
```

```python
# 
# Create a scatterplot graph using seaborn showing the relationship between three columns
sns.scatterplot(data=iris_df, x="sepal_length", y="sepal_width", hue="species")
# 
# Give a Title to the scatterplot graph
plt.title("Iris Sepal Length vs Width")
# 
# Draw the graph
plt.show()
```
